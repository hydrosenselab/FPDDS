## Libraries

```python
import os, sys
import xarray as xr
import matplotlib.pyplot as plt

# Project-Specific package
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
import pylisf as pl 
```

## User-entry

```python
parfile='/home/civil/faculty/msaharia/LDAS/01-testcases/IND/lisinputind.nc' #LIS parameter file
```

## Data preparation

```python
pardat=xr.open_dataset(parfile)
```

```python
pardat=pl.reformat_LIS_paramfile(pardat)
pardat
```

```python
parvarlist = pl.nc_varlist(pardat) #List of all netCDF variables
unwantedvars = ['lat', 'lon', 'lat_b', 'lon_b', 'time'] #Variables that you want to remove the above list
parvarlist = [e for e in parvarlist if e not in unwantedvars]
```

## Visualization

```python
fig, axs = plt.subplots(21, 2, figsize=(15,150), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .2, wspace= 0.2)

axs = axs.ravel()
for count, value in enumerate(parvarlist):
    pardat[value].plot(ax=axs[count])
    axs[count].set_title(str(value))
    axs[count].xaxis.label.set_visible(False)
    axs[count].yaxis.label.set_visible(False)
```
