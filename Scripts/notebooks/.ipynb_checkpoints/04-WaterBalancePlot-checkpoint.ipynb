# Purpose
Plot Water Balance using LIS outputs

## Libraries

```python
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Project-Specific package
import os, sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import pylisf as pl 
```

## User-defined

```python
lsmdat=xr.open_mfdataset('../../../02-ILDAS/OUTPUT/EXP000/SURFACEMODEL/*/*HIST*', 
                      combine='by_coords')
routedat=xr.open_mfdataset('../../../02-ILDAS/OUTPUT/EXP000/ROUTING/*/*HIST*', 
                      combine='by_coords')
latsel=15.55
lonsel=97.75

lsmvars={
    'Rainf_tavg': 'Rain',
    'SWE_tavg': 'SWE'
}

routevars={
    'FloodedFrac_tavg': 'Flooded Fraction',
    'RiverDepth_tavg':'River Depth',
    'SWS_tavg': 'Surface Water Storage'
}
```

## Read data

```python
lsmdat=pl.reformat_LIS_output(lsmdat)
routedat=pl.reformat_LIS_output(routedat)
```

## Plotting

```python
numfigs=len(lsmvars)+len(routevars)+1
figheight=numfigs*2

fig, axes = plt.subplots(ncols=1, nrows=numfigs, figsize=(9,figheight), tight_layout=True)
fig.suptitle("LIS Simulation over India \n Latitude=" + str(latsel) + ' and longitude=' + str(lonsel), size=13, x=0.65, y=1.03)
myFmt = mdates.DateFormatter('%b-%y')

# Plotting LSM variables 
for index, (key, value) in enumerate(lsmvars.items()):
    #print(index, key, value) #0, Rainf_tavg, Rain
    lsmdat[key].sel(lat=latsel, lon=lonsel, method='nearest').plot(ax=axes[index])
    axes[index].set_title('')
    axes[index].xaxis.set_major_formatter(mdates.DateFormatter('%b-%y'))
    axes[index].set_ylabel(lsmvars[key] + '\n' + 
                           '(' + lsmdat[key].units + ')', 
                           rotation='horizontal', horizontalalignment='right')
    axes[index].yaxis.set_label_coords(-0.11, 0.4)   
    axes[index].tick_params('x', labelrotation=0)
    
# Plotting routing variables 
for index, (key, value) in enumerate(routevars.items()):
    newindex=index+len(lsmvars)
    routedat[key].sel(lat=latsel, lon=lonsel, method='nearest').plot(ax=axes[newindex])
    axes[newindex].set_title('')
    axes[newindex].xaxis.set_major_formatter(mdates.DateFormatter('%b-%y'))
    axes[newindex].set_ylabel(routevars[key] + '\n' + 
                           '(' + routedat[key].units + ')', 
                           rotation='horizontal', horizontalalignment='right')
    axes[newindex].yaxis.set_label_coords(-0.11, 0.4)   
    axes[newindex].tick_params('x', labelrotation=0) 
    
# Streamflow Plot
finalindex=len(lsmvars)+len(routevars)
routedat['Streamflow_tavg'].sel(lat=latsel, lon=lonsel, method='nearest').plot(ax=axes[finalindex])
axes[finalindex].set_title('')
axes[finalindex].xaxis.set_major_formatter(mdates.DateFormatter('%b-%y'))
axes[finalindex].set_ylabel('Streamflow'+ '\n' + 
                           '(' + routedat['Streamflow_tavg'].units + ')', 
                            rotation='horizontal', horizontalalignment='right')
axes[finalindex].yaxis.set_label_coords(-0.11, 0.4)   
axes[finalindex].tick_params('x', labelrotation=0) 
```

```python
fig.savefig('wbplot.png')
```
