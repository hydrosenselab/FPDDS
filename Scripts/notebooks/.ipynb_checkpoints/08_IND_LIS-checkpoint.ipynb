```python
import itertools
def split_by_chunks(dataset):
    chunk_slices = {}
    for dim, chunks in dataset.chunks.items():
        slices = []
        start = 0
        for chunk in chunks:
            if start >= dataset.sizes[dim]:
                break
            stop = start + chunk
            slices.append(slice(start, stop))
            start = stop
        chunk_slices[dim] = slices
    for slices in itertools.product(*chunk_slices.values()):
        selection = dict(zip(chunk_slices.keys(), slices))
        yield dataset[selection]
```

```python
datasets = list(split_by_chunks(routeclip))
```

```python
def create_filepath(ds, prefix='ROUTE', root_path="../data/output/LIS_IND/ROUTE_IND"):
    """
    Generate a filepath when given an xarray dataset
    """
    start = ds.time.data[0]
    end = ds.time.data[-1]
    filepath = f'{root_path}/{prefix}_{end}.nc'
    return filepath
```
