```python
error_df1 = error_df[error_df['NSE'] >= -5 ]
# error_df1 = error_df1[error_df1['MAE'] <= 1000 ]
error_df1 = error_df1[error_df1['PBIAS'] <= 600 ]
```

```python
error_df1.describe()
```

```python
error_df2 = error_df1['NMAE'].quantile(0.75)
```

```python
error_df2
```

```python
metric_list = ['PBIAS','RRMSE', 'NSE', 'KGE','CORR', 'NMAE']
metric_names = ['Percent Bias (PBIAS)','Relative Root Mean Square Error (RRMSE)',
                'Nash-Sutcliffe Efficiency (NSE)','Kling-Gupta Efficiency (KGE)',
               'Correlation Coefficient (CORR)','Normalised Mean Absolute Error (NMAE)']
metric_vmax = [100,4.7175,None,None,None,2.06]
metric_vmin = [-100,None,-1,-1,0.3,None]
metric_cmap = ['hsv','jet','jet_r','jet_r','jet_r','jet']
metric_extend = ['both','max','min','min','min','max']
```

```python
fig, axes = plt.subplots(3, 2, figsize=(8.6, 11),dpi = 120,constrained_layout=True )
font_size = 10
for i, ax in enumerate(axes.flatten()):
    density = stats.gaussian_kde(error_df1[metric_list[i]])
    median = round(error_df1[metric_list[i]].median(),3)
    n, x, _ = ax.hist(error_df1[metric_list[i]], bins=50,alpha=0.5,edgecolor='black', linewidth=0.8, density=True)
    ax.plot(x, density(x))
    ax.set_ylabel('Density')
    ax.set_xlabel('Value')
    ax.set_title(metric_list[i]+' (National Median): '+str(median),loc='left' , fontsize=font_size)
plt.plot()
plt.savefig('../output/IND_density.png',format = 'png', dpi=300)
plt.close
```

```python
lat = error_df1['Latitude'].values
lon = error_df1['Longitude'].values
num_colors = 50
marker_size = 10
fig, axes = plt.subplots(3, 2, figsize=(8.6, 11),dpi = 120,constrained_layout=True )
font_size = 10
for i, ax in enumerate(axes.flatten()):
    cmap = plt.get_cmap(metric_cmap[i], num_colors)
    cmap.set_under('black')
    cmap.set_over('black')
    metric = error_df1[metric_list[i]].values
    ind_map.plot(ax=ax,color = 'white',edgecolor = (0.5,0.5, 0.5, 1))
    im = ax.scatter(lon, lat,
                 c= metric,
                 cmap=cmap, alpha=0.8,s = marker_size,edgecolor='black',linewidth=0.5, vmin=metric_vmin[i], vmax=metric_vmax[i])
    ax.set_title(metric_names[i],loc='left', fontsize=font_size)
    ax.set_xticks([])
    ax.set_yticks([])
    # ax.tick_params(axis='both', labelsize=8)
    divider = axes_grid1.make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar = fig.colorbar(im, cax=cax,extend=metric_extend[i])
plt.plot()
plt.savefig('../output/IND_metrics_a.png',format = 'png', dpi=300)
plt.close
```

```python
lsm_monthly = xr.open_dataset('../LIS_Data/EXP002/LSM_IND_monthly.nc')
lsm_yearly = xr.open_dataset('../LIS_Data/EXP002/LSM_IND_yearly.nc')
route_monthly = xr.open_dataset('../LIS_Data/EXP002/route_IND_monthly.nc')
route_yearly = xr.open_dataset('../LIS_Data/EXP002/route_IND_yearly.nc')
```

```python
lsm_annual = lsm_yearly.mean(dim='year')
route_annual = route_yearly.mean(dim='year')
```

```python
lsm_annual = lsm_annual.assign(soil_L1=(lsm_annual['SoilMoist_tavg'].sel(SoilMoist_profiles = 0))*(1000/100))
lsm_annual = lsm_annual.assign(soil_L2=(lsm_annual['SoilMoist_tavg'].sel(SoilMoist_profiles = 1))*(1000/300))
lsm_annual = lsm_annual.assign(soil_L3=(lsm_annual['SoilMoist_tavg'].sel(SoilMoist_profiles = 2))*(1000/600))
lsm_annual = lsm_annual.assign(soil_L4=(lsm_annual['SoilMoist_tavg'].sel(SoilMoist_profiles = 3))*(1000/1000))
lsm_annual = lsm_annual.assign(annual_precip=(lsm_annual['TotalPrecip_tavg'])*(86400*30*12))
lsm_annual = lsm_annual.assign(annual_evap=(lsm_annual['Evap_tavg'])*(86400*30*12))
lsm_annual = lsm_annual.assign(annual_surface_q=(lsm_annual['Qs_tavg'])*(86400*30*12))
lsm_annual = lsm_annual.assign(annual_subsurface_q=(lsm_annual['Qsb_tavg'])*(86400*30*12))
lsm_annual = lsm_annual.assign(annual_tws=(lsm_annual['TWS_tavg'])*(86400*30*12))
```

```python
lsm_vars = ['annual_precip','annual_evap','soil_L1','soil_L2','soil_L3','soil_L4']
lsm_name = ['Mean Precipitation (mm/yr)','Mean Evapotranspiration (mm/yr)','Mean Soil Moisture (mm)','Mean Soil Moisture (mm)','Mean Soil Moisture (mm)','Mean Soil Moisture (mm)']
lsm_inset = ["2004-2014","2004-2014",'Layer: 0-100 mm\n2004-2014','Layer: 100-400 mm\n2004-2014','Layer:400-1000 mm\n2004-2014','Layer: 1000-2000 mm\n2004-2014']
lsm_sel = [None,None,0,1,2,3,None]
```

```python
fig, axes = plt.subplots(3, 2, figsize=(8,14),dpi = 120,constrained_layout=False )
font_size = 10
plt.style.use('seaborn-white')
for i, ax in enumerate(axes.flatten()):
    im=xr.plot.imshow(lsm_annual[lsm_vars[i]], 
                      cmap="jet",add_colorbar=False,ax=ax)
    ax.set_title(lsm_name[i],size = 9,loc='left')
    ax.text(0.8, 0.95,lsm_inset[i],
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax.transAxes,size=7.5)
    ax.set_title("",size = 9)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylabel('')
    ax.set_xlabel('')
    # ax.tick_params(axis='both', labelsize=8)
    divider = axes_grid1.make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar = fig.colorbar(im, cax=cax)
    # ax.set_facecolor('white')
plt.savefig('../../output/IND_LDAS.png',format = 'png', dpi=300)
plt.show()
```
