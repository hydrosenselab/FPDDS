```python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.colors as mcl
from mpl_toolkits import axes_grid1
from shapely.geometry import Point, shape
import fiona 
import shapely
import tqdm
import scipy.stats as stats
import numpy as np
import pandas as pd
import os,glob
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from reportlab.lib.pagesizes import portrait
from reportlab.platypus import Image
from reportlab.platypus import Table
from reportlab.lib import colors
```

```python
ind_map = gpd.read_file('../../LIS_Data/maps-master/States/Admin2.shp')
```

```python
error_df = pd.read_csv('../../LIS_Data/EXP002/error_metric_new.csv')
```

```python
## Build the list of basin files
# basin_files = ['Basin_Krishna','Basin_Narmada','Basin_North_western','Basin_Cauvery_Eastern','Basin_Mahanadi_Subarnarekha','Basin_Godavari']
# basin_files = ['Basin_Narmada']
# basin_params = {'Basin_Krishna':[8,8],'Basin_Narmada':[7,10],'Basin_North_western':[10,7],'Basin_Cauvery_Eastern':[8,8],'Basin_Mahanadi_Subarnarekha':[8,8],'Basin_Godavari':[8,8]}
basin_params = {'Basin_N_T_G_S_M':[12,12],'Basin_C_K_W_E':[18,10],'Basin_North_western':[16,10]}
```

```python
error_df1 = error_df[error_df['NSE'] >= -5 ]
error_df1 = error_df1[error_df1['MAE'] <= 1000 ]
error_df1 = error_df1[error_df1['PBIAS'] <= 600 ]
```

```python
error_df1['NMAE'] = error_df1['MAE']/error_df1['obs_avg']
```

```python
error_df1['NMAE'].quantile(.75)
```

```python
metric_list = ['PBIAS','RRMSE', 'NSE', 'KGE','CORR', 'NMAE']
# metric_list = ['PBIAS']
metric_names = ['Percent Bias (PBIAS)','Relative Root Mean Square Error (RRMSE)',
                'Nash-Sutcliffe Efficiency (NSE)','Kling-Gupta Efficiency (KGE)',
               'Correlation Coefficient (CORR)','Normalised Mean Absolute Error (NMAE)']
metric_vmax = [100,6,None,None,None,2.06]
metric_vmin = [-100,None,-1,-1,0.3,None]
metric_cmap = ['hsv','jet_r','jet_r','jet_r','jet_r','jet']
metric_extend = ['max','max','min','min','min','max']
```

```python
error_df1.reset_index(inplace=True)
```

```python
for item in basin_params.keys():
    fc = fiona.open('../../shapefiles/india_basins/'+item+'_Dissolved.shp')
    basin_map = gpd.read_file('../../shapefiles/india_basins/'+item+'.shp')
    basin_points = pd.DataFrame()
    shapefile_record = fc.next()
    shape = shapely.geometry.asShape( shapefile_record['geometry'] )
    for k in range(0,len(error_df1)) :
        x = error_df1.loc[k]['Longitude']
        y = error_df1.loc[k]['Latitude']
        point = shapely.geometry.Point(x,y) # longitude, latitude
        # Alternative: if point.within(shape)
        if shape.contains(point):
            basin_points = basin_points.append(error_df1.loc[k])
    fig = plt.figure()
    num_colors = 50
#     cmap = plt.get_cmap('jet_r', num_colors)
#     cmap.set_under('magenta')
#     cmap.set_over('magenta')
    lat = basin_points['Latitude'].values
    lon = basin_points['Longitude'].values
    ########################################################
    ax1 = fig.add_subplot(3,2,1)
    i = 0
    cmap = plt.get_cmap(metric_cmap[i], num_colors)
    cmap.set_under('black')
    cmap.set_over('black')
    metric = basin_points[metric_list[i]].values
    median = round(basin_points[metric_list[i]].median(),3)
    basin_map.plot(ax=ax1,color = 'white',edgecolor = (0.5,0.5, 0.5, 1))
    im = plt.scatter(lon, lat,
                 c= metric,
                 cmap=cmap, alpha=1, label="Gauge",edgecolors='b',vmin=metric_vmin[i], vmax=metric_vmax[i])
    plt.title(metric_names[i]+'\nMedian: '+str(median),loc='left', fontsize=12)
    # ax.tick_params(axis='both', labelsize=8)
    divider = axes_grid1.make_axes_locatable(ax1)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar = fig.colorbar(im, cax=cax,extend=metric_extend[i])
    # cbar.ax.tick_params(labelsize=8) 
    ##########################################################
    ##########################################################
    ax2 = fig.add_subplot(3,2,2)
    i = 1
    cmap = plt.get_cmap(metric_cmap[i], num_colors)
    cmap.set_under('black')
    cmap.set_over('black')
    metric = basin_points[metric_list[i]].values
    median = round(basin_points[metric_list[i]].median(),3)
    basin_map.plot(ax=ax2,color = 'white',edgecolor = (0.5,0.5, 0.5, 1))
    im = plt.scatter(lon, lat,
                 c= metric,
                 cmap=cmap, alpha=1, label="Gauge",edgecolors='b',vmin=metric_vmin[i], vmax=metric_vmax[i])
    plt.title(metric_names[i]+'\nMedian: '+str(median),loc='left', fontsize=12)
    # ax.tick_params(axis='both', labelsize=8)
    divider = axes_grid1.make_axes_locatable(ax2)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar = fig.colorbar(im, cax=cax,extend=metric_extend[i])
    # cbar.ax.tick_params(labelsize=8) 
    ##########################################################
    ax3 = fig.add_subplot(3,2,3)
    i = 2
    cmap = plt.get_cmap(metric_cmap[i], num_colors)
    cmap.set_under('black')
    cmap.set_over('black')
    metric = basin_points[metric_list[i]].values
    median = round(basin_points[metric_list[i]].median(),3)
    basin_map.plot(ax=ax3,color = 'white',edgecolor = (0.5,0.5, 0.5, 1))
    im = plt.scatter(lon, lat,
                 c= metric,
                 cmap=cmap, alpha=1, label="Gauge",edgecolors='b',vmin=metric_vmin[i], vmax=metric_vmax[i])
    plt.title(metric_names[i]+'\nMedian: '+str(median),loc='left', fontsize=12)
    # ax.tick_params(axis='both', labelsize=8)
    divider = axes_grid1.make_axes_locatable(ax3)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar = fig.colorbar(im, cax=cax,extend=metric_extend[i])
    # cbar.ax.tick_params(labelsize=8) 
    ##########################################################
    ax4 = fig.add_subplot(3,2,4)
    i = 3
    cmap = plt.get_cmap(metric_cmap[i], num_colors)
    cmap.set_under('black')
    cmap.set_over('black')
    metric = basin_points[metric_list[i]].values
    median = round(basin_points[metric_list[i]].median(),3)
    basin_map.plot(ax=ax4,color = 'white',edgecolor = (0.5,0.5, 0.5, 1))
    im = plt.scatter(lon, lat,
                 c= metric,
                 cmap=cmap, alpha=1, label="Gauge",edgecolors='b',vmin=metric_vmin[i], vmax=metric_vmax[i])
    plt.title(metric_names[i]+'\nMedian: '+str(median),loc='left', fontsize=12)
    # ax.tick_params(axis='both', labelsize=8)
    divider = axes_grid1.make_axes_locatable(ax4)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar = fig.colorbar(im, cax=cax,extend=metric_extend[i])
    # cbar.ax.tick_params(labelsize=8) 
    ##########################################################
    ax5 = fig.add_subplot(3,2,5)
    i = 4
    cmap = plt.get_cmap(metric_cmap[i], num_colors)
    cmap.set_under('black')
    cmap.set_over('black')
    metric = basin_points[metric_list[i]].values
    median = round(basin_points[metric_list[i]].median(),3)
    basin_map.plot(ax=ax5,color = 'white',edgecolor = (0.5,0.5, 0.5, 1))
    im = plt.scatter(lon, lat,
                 c= metric,
                 cmap=cmap, alpha=1, label="Gauge",edgecolors='b',vmin=metric_vmin[i], vmax=metric_vmax[i])
    plt.title(metric_names[i]+'\nMedian: '+str(median),loc='left', fontsize=12)
    # ax.tick_params(axis='both', labelsize=8)
    divider = axes_grid1.make_axes_locatable(ax5)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar = fig.colorbar(im, cax=cax,extend=metric_extend[i])
    # cbar.ax.tick_params(labelsize=8) 
    ##########################################################
    ax6 = fig.add_subplot(3,2,6)
    i = 5
    cmap = plt.get_cmap(metric_cmap[i], num_colors)
    cmap.set_under('black')
    cmap.set_over('black')
    metric = basin_points[metric_list[i]].values
    median = round(basin_points[metric_list[i]].median(),3)
    basin_map.plot(ax=ax6,color = 'white',edgecolor = (0.5,0.5, 0.5, 1))
    im = plt.scatter(lon, lat,
                 c= metric,
                 cmap=cmap, alpha=1, label="Gauge",edgecolors='b',vmin=metric_vmin[i], vmax=metric_vmax[i])
    plt.title(metric_names[i]+'\nMedian: '+str(median),loc='left', fontsize=12)
    # ax.tick_params(axis='both', labelsize=8)
    divider = axes_grid1.make_axes_locatable(ax6)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar = fig.colorbar(im, cax=cax,extend=metric_extend[i])
    # cbar.ax.tick_params(labelsize=8) 
    ########################################################
    fig.set_figheight(basin_params[item][0])
    fig.set_figwidth(basin_params[item][1])
    plt.plot()
    plt.savefig('../../output/'+ item + '.svg',format = 'svg', dpi=300)
    plt.close
```
