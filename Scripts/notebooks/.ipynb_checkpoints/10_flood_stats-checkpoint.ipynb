```python
def flood_metrics(gauge_list,n):
    filename = os.path.basename(gauge_list[n])
    s_name = filename.rsplit('_', 2)           # for ne_monsoon, cutoff at 2, otherwise 1
    station = pd.read_csv(gauge_list[n])
    station.dropna(inplace=True)
    streamflow = station[['Date','Streamflow (cumecs)','Streamflow_tavg']].copy()
    streamflow = streamflow.rename(columns={'Streamflow (cumecs)': 'Obs_Flow(cumecs)','Streamflow_tavg':'Sim_Flow(cumecs)'})
    streamflow['Obs_F'] = np.nan
    streamflow['Sim_F'] = np.nan
    streamflow['Alarm'] = np.nan
    streamflow.reset_index(drop=True,inplace=True)
    for i in range (0,len(streamflow)):
        if streamflow.loc[i,'Obs_Flow(cumecs)'] >= streamflow['Obs_Flow(cumecs)'].quantile(qt):
            streamflow.loc[i,'Obs_F'] = 'Flood'
        if streamflow.loc[i,'Obs_Flow(cumecs)'] < streamflow['Obs_Flow(cumecs)'].quantile(qt):
            streamflow.loc[i,'Obs_F'] = 'No Flood'
        if streamflow.loc[i,'Sim_Flow(cumecs)'] >= streamflow['Obs_Flow(cumecs)'].quantile(qt):
            streamflow.loc[i,'Sim_F'] = 'Flood'
        if streamflow.loc[i,'Sim_Flow(cumecs)'] < streamflow['Obs_Flow(cumecs)'].quantile(qt):
            streamflow.loc[i,'Sim_F'] = 'No Flood'
        if streamflow.loc[i,'Obs_F'] == 'Flood' and streamflow.loc[i,'Sim_F'] == 'Flood':
            streamflow.loc[i,'Alarm'] = 'H'
        if streamflow.loc[i,'Obs_F'] == 'Flood' and streamflow.loc[i,'Sim_F'] == 'No Flood':
            streamflow.loc[i,'Alarm'] = 'M'
        if streamflow.loc[i,'Obs_F'] == 'No Flood' and streamflow.loc[i,'Sim_F'] == 'Flood':
            streamflow.loc[i,'Alarm'] = 'FA'
        if streamflow.loc[i,'Obs_F'] == 'No Flood' and streamflow.loc[i,'Sim_F'] == 'No Flood':
            streamflow.loc[i,'Alarm'] = 'CN'
    streamflow.to_csv(path+s_name[0]+'-flood.csv')
```

```python
flood_list = sorted(glob.glob(path+'*-flood.csv'))
flood_stats = pd.DataFrame(columns=['GaugeID','H','M','FA','CN','HR','FAR','CSI','POFD','BIAS','Accuracy'])   
for n in tqdm (range(0,len(flood_list))):
    filename = os.path.basename(flood_list[n])
    s_name = filename.rsplit('-', 1)
    streamflow = pd.read_csv(flood_list[n])
    flood_stats.loc[n,'GaugeID'] = s_name[0]
    flood_stats.loc[n,'H'] = len(streamflow.query("Alarm == 'H'"))
    flood_stats.loc[n,'M'] = len(streamflow.query("Alarm == 'M'"))
    flood_stats.loc[n,'FA'] = len(streamflow.query("Alarm == 'FA'"))
    flood_stats.loc[n,'CN'] = len(streamflow.query("Alarm == 'CN'"))
    try :
        flood_stats.loc[n,'HR'] = (flood_stats.loc[n,'H']/(flood_stats.loc[n,'H']+flood_stats.loc[n,'M']))*100
        flood_stats.loc[n,'FAR'] = (flood_stats.loc[n,'FA']/(flood_stats.loc[n,'H']+flood_stats.loc[n,'FA']))*100
        flood_stats.loc[n,'CSI'] = (flood_stats.loc[n,'H']/(flood_stats.loc[n,'H']+flood_stats.loc[n,'FA']+flood_stats.loc[n,'M']))*100
        flood_stats.loc[n,'POFD'] = (flood_stats.loc[n,'FA']/(flood_stats.loc[n,'FA']+flood_stats.loc[n,'CN']))*100
    except Exception as e :
        continue
flood_stats.to_csv(path+'flood_metadata.csv')
```
