```python
import geopandas as gpd
import scipy.stats as stats
import numpy as np
import pandas as pd
import os,sys,glob
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.colors as mcl
from mpl_toolkits import axes_grid1
import seaborn as sns
from IPython.display import HTML
import xarray as xr
import rioxarray
import rasterio as rio
from tqdm import tqdm
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import config
import gc
```

```python
drainagearea = pd.read_csv('../../final_results/IWM-Validation-EXP002/rawdata/hydro_signatures/drainage_area.csv')
drainagearea.set_index('GaugeID',inplace=True)
gauge_list = sorted(glob.glob('/home/civil/phd/cez198621/projects/IWM_verification/final_results/IWM-Validation-EXP002/rawdata/gaugestreamflow/merged/all_year/IWM*.csv'))
```

```python
column_list = ['GaugeID','Latitude','Longitude','Mean Annual Flow (m3/s)',
                                  'Mean Annual Monsoon Flow (m3/s)','Mean Daily Flow (m3/s)',
                                  'Mean Daily Baseflow (m3/s)','Base Flow Index',
                                  'Slope of FDC','Low Flow (m3/s)',
                                  'High Flow (m3/s)',
                                  'Annual HF Frequency (Days)','Annual LF Frequency (Days)','Runoff Ratio','Streamflow Elasticity']
hydro_sig = pd.DataFrame(columns=column_list).set_index('GaugeID')
flow_var = 'Streamflow_tavg'  # Select observed or modeled streamflow
for i in tqdm(range (0,len(gauge_list))):
    try:
        filename = os.path.basename(gauge_list[i])
        s_name = filename.rsplit('-', 1)           # for ne_monsoon, cutoff at 2, otherwise 1
        station = pd.read_csv(gauge_list[i])
        streamflow = station[['Date','Streamflow (cumecs)','Streamflow_tavg','TotalPrecip_tavg']].copy()
        streamflow['Streamflow (cumecs)'] = streamflow['Streamflow (cumecs)'].replace(0, np.nan)  # Drop all "0" observed streamflow rows 
        streamflow.dropna(inplace=True)
        streamflow.set_index('Date',inplace=True)
        streamflow.index = pd.to_datetime(streamflow.index)
        monsoon = streamflow[streamflow.index.map(lambda t: t.month in [6,7,8,9])] 
        streamflow['Runoff (mm/s)'] = (streamflow[flow_var]/(drainagearea.loc[s_name[0],'Drainage_Area']*1000000))*1000  # Calculate daily runoff in mm/s
        series = streamflow.loc[:,flow_var] #Generate a timeseries for base flow index calculations
        q_low = streamflow[flow_var].quantile(0.05)
        q_high = streamflow[flow_var].quantile(0.95)
        fdc_slope = config.fdc_slope(streamflow[flow_var])
        q_annual_mean = config.annual_mean(streamflow[flow_var])
        q_monsoon_mean = config.annual_mean(monsoon[flow_var])
        hq_freq,lq_freq = config.flow_freq(streamflow[flow_var])
        b_flow = config.IHmethod(series, block_length=5, tp=0.9, interp_semilog=True)
        mean_bflow = b_flow['QB'].mean()
        q_mean = series.mean()
        bfi = mean_bflow/q_mean
        lat = drainagearea.loc[s_name[0],'Latitude']
        lon = drainagearea.loc[s_name[0],'Longitude']
        d_area = drainagearea.loc[s_name[0],'Drainage_Area'] 
        r_ratio,elasticity = config.streamflow_elasticity(streamflow['Runoff (mm/s)'],streamflow['TotalPrecip_tavg'])
        new_row = pd.DataFrame([[s_name[0],lat,lon,q_annual_mean,q_monsoon_mean,q_mean,mean_bflow,bfi,fdc_slope,q_low,q_high,hq_freq,lq_freq,r_ratio,elasticity]],columns=column_list).set_index('GaugeID')
        hydro_sig = hydro_sig.append(new_row)
    except Exception as e:
        print (e)
        continue
```

```python
hydro_sig.to_csv('../../final_results/IWM-Validation-EXP002/rawdata/hydro_signatures/modeled_hydro_signatures.csv')
```

```python
observed_sig = pd.read_csv('../../final_results/IWM-Validation-EXP002/rawdata/hydro_signatures/observed_hydro_signatures.csv').set_index('GaugeID')
modeled_sig  = pd.read_csv('../../final_results/IWM-Validation-EXP002/rawdata/hydro_signatures/modeled_hydro_signatures.csv').set_index('GaugeID')
```

```python
signatures = ['Mean Annual Flow (m3/s)',
              'Mean Annual Monsoon Flow (m3/s)','Mean Daily Flow (m3/s)',
              'Mean Daily Baseflow (m3/s)','Base Flow Index',
              'Slope of FDC','Low Flow (m3/s)','High Flow (m3/s)','Runoff Ratio','Streamflow Elasticity']
```

```python
hydro_rsquare  = pd.DataFrame(columns=['S.No','Signature','R_square'])
for i in range (0,len(signatures)):
    obs_sig = observed_sig[signatures[i]]
    sim_sig = modeled_sig[signatures[i]]
    obs_sig.dropna(inplace=True)
    sim_sig.dropna(inplace=True)
    sn = i+1
    name = signatures[i]
    val = round(config.rsquared(obs_sig,sim_sig),3)
    hydro_rsquare.loc[i] = pd.Series({'S.No':sn, 'Signature':name, 'R_square':val})
```
