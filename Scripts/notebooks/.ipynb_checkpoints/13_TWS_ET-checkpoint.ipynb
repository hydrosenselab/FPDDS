```python
for n in tqdm(range (0,(len(batch_index)))):
    lsm = xr.open_mfdataset(lsm_files[batch_index[n]:batch_index1[n]],combine='by_coords',parallel=True)
    lsm1 = config.reformat_LIS_output(lsm)
    #lsm = lsm.assign(total_rainfall= (lsm['TotalPrecip_tavg']*86400))
    #lsm1 = lsm['total_rainfall']
    #lsm1 = lsm1.assign(total_rainfall= (lsm1['TotalPrecip_tavg']*86400))
    lsm1  = lsm1.chunk({'time':366})
    lsm1 = lsm1.compute()
    lsm1.rio.set_spatial_dims(x_dim="lon", y_dim="lat", inplace=True)
    lsm1.rio.write_crs("epsg:4326", inplace=True)
    lsm_clip = lsm1.rio.clip(mask.geometry.apply(mapping), mask.crs, drop=True)
    lsm_monthly = lsm_clip.resample(time='1M').sum()
    #lsm_yearly = lsm1.resample(time='1Y').sum()
    lsm_monthly.to_netcdf('/home/civil/phd/cez218606/LISF1/Results/MERRA2/lsm_monthly1/lsm_merra2_monthly_'+str(n)+'.nc')
    del lsm1
    del lsm_clip
    del lsm
    del lsm_monthly
    #del lsm_yearly
    gc.collect()
```
