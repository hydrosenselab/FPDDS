```python
def covariance(x, y, dims=None):
    return xr.dot(x - x.mean(dims), y - y.mean(dims), dims=dims) / x.count(dims)

def corrrelation(x, y, dims=None):
    return covariance(x, y, dims) / (x.std(dims) * y.std(dims))

def rmse(x, y, dims=None):
    return np.sqrt(((x - y) ** 2).mean(dims))

def bias(x, y, dims=None):
    return ((x - y).mean(dims))


def KGE(x, y, dims=None):
    return  (1- np.sqrt((1-corrrelation(x, y, dims))**2 + (1-(x.mean(dims)/y.mean(dims)))**2 + (1-(x.std(dims)/y.std(dims)))**2))
```

```python
lsmsm = lsm_anomalies.soil_moisture
esasm = esa_anomalies.sm
```
