```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:


import os, sys, glob
import time
import subprocess
import multiprocessing
from multiprocessing import Pool
import shutil
for j in range(0,35):
    # Base shell command structure for each model run

    def run_model(model_id):
        shell_script = f"""
        cd /home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/
        # Load necessary modules
        module purge
        module load lib/intel/2019/eccodes/2.18.0
        module load lib/intel/2019/esmf/8.0.0
        module load lib/intel/2019/hdfeos/2.20v1.10
        module swap suite/intel/parallelStudio/2019 suite/intel/parallelStudio/2018
        # Execute the primary job with mpirun
        time -p mpirun -np 4 /home/civil/phd/cez218606/LISF1/LISF-7.4/LISF/lis/LIS -f /home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/LIS_IMD_{model_id}

        # Execute final Python script
        cd /home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/output/pyLISF/notebooks/
        ~/anaconda3/envs/geo_env/bin/python /home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/output/pyLISF/notebooks/calibration_KGE_new{model_id}.py


       """

        # Define unique log file paths for each model run
        output_log = f"/home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/logs/output_mp_{model_id}.txt"
        error_log = f"/home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/logs/error_mp_{model_id}.txt"

        # Execute the shell script using subprocess and direct output to log files
        with open(output_log, "w") as out, open(error_log, "w") as err:
            process = subprocess.run(shell_script, shell=True, executable="/bin/bash", stdout=out, stderr=err)

        # Check and return job status
        if process.returncode == 0:
            print(f"Model run {model_id} completed successfully.")
        else:
            print(f"Model run {model_id} failed with return code: {process.returncode}")
        return process.returncode

    # Number of parallel model runs
    model_ids = [1, 2, 3, 4]

    if __name__ == "__main__":
        # Initialize a pool with four workers
        with Pool(processes=4) as pool:
            # Map the run_model function to each model_id in model_ids
            results = pool.map(run_model, model_ids)

        # Check if all jobs completed successfully
        if all(code == 0 for code in results):
            print("All model runs completed successfully.")
        else:
            print("Some model runs failed.")




    # Function to execute the job with a unique ID
    def run_job(job_id):
        shell_script = f"""
        # Execute Python script to save best run
        cd /home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/output/pyLISF/notebooks/
        ~/anaconda3/envs/geo_env/bin/python /home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/output/pyLISF/notebooks/save_best_run.py

        # Ostrich setup
        cd /home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/NoahMP_InputFiles/
        cp ostIn_Parallel.txt ostIn.txt
        echo 0 > Counter.txt
        module load lib/intel
        export I_MPI_PMI_LIBRARY=/usr/lib64/libpmi.so
        mpirun -n 4 /home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/Linux/intelmpi/5.1.1/OstrichMPI

        # Execute final Python script
        cd /home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/output/pyLISF/notebooks/
        ~/anaconda3/envs/geo_env/bin/python /home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/output/pyLISF/notebooks/multiplire.py
        """

        # Define log file paths for the job
        output_log = f"/home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/logs/output_mo_{job_id}.txt"
        error_log = f"/home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/logs/error_mo_{job_id}.txt"

        # Run the shell script and log output and errors
        with open(output_log, "w") as out_log, open(error_log, "w") as err_log:
            subprocess.run(shell_script, shell=True, stdout=out_log, stderr=err_log)
    # Run the job
    run_job(job_id=1)  # Specify the job ID or modify as needed


    shutil.rmtree('/home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/output/LSM_IMD1/ROUTING/200312')
    shutil.rmtree('/home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/output/LSM_IMD2/ROUTING/200312') 
    shutil.rmtree('/home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/output/LSM_IMD3/ROUTING/200312') 
    shutil.rmtree('/home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/output/LSM_IMD4/ROUTING/200312')
    for i in range(0, 4):
        path1 = f'/home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/NoahMP_InputFiles/param_save/er{j}/mod{i}'
        path2 = f'/home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/NoahMP_InputFiles/mod{i}'
        path3 = f'/home/civil/phd/cez218606/Auto_calibration/Final/Ganga_PDDS/output/error_metrix/'

        os.makedirs(path1, exist_ok=True)

        # Copy all files from path2 to path1
        for file_name in os.listdir(path2):
            full_file_name = os.path.join(path2, file_name)
            if os.path.isfile(full_file_name):
                shutil.copy(full_file_name, path1)

        # Copy all files from path3 to path1
        for file_name in os.listdir(path3):
            full_file_name = os.path.join(path3, file_name)
            if os.path.isfile(full_file_name):
                shutil.copy(full_file_name, path1)
    print('j iteration is completion')
```
