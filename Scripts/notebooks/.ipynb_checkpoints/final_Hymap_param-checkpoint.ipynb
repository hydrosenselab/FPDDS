```python
def covariance(x, y):
    return np.dot(x - x.mean(), y - y.mean()) / x.count()

def corrrelation(x, y):
    return covariance(x, y) / (x.std() * y.std())

def R2(x, y):
    return (corrrelation(x, y))**2
```

```python
x1 = grun1['ET']
y1 = uncal1['ET']
df1 = pd.DataFrame({"x1":x1,"y1":y1}).dropna()
uncal_c_r = R2(df1['x1'],df1['y1'])     
uncal_c_k = config.kge(x1,y1,return_all="True")[1]
uncal_c_n = config.nashsutcliffe(x1,y1)

x1 = grun1['ET']
y1 = cal1['ET']
df1 = pd.DataFrame({"x1":x1,"y1":y1}).dropna()
ical_c_r =R2(df1['x1'],df1['y1'])  
ical_c_k = config.kge(x1,y1,return_all="True")[0]
ical_c_n = config.nashsutcliffe(x1,y1)
```

```python
fig = plt.subplots(1, 1, figsize=(8,4),dpi = 300,constrained_layout=False ,facecolor='white')
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams['axes.facecolor'] = 'white'


plt.ylim([0,7.5])
grun1['ET'].plot(label='MODIS  [KGE, $R^2$, NSE]',color = 'gray',linestyle="--",linewidth=1)
uncal1['ET'].plot(color = 'steelblue',linewidth=0.8,label='unCalib  ['+str("%.2f" %(uncal_c_k))+','+str("%.2f" %(uncal_c_r))+','+str("%.2f" %(uncal_c_n))+']',linestyle="-.")
cal1['ET'].plot(color = 'firebrick',linewidth=0.8, label='Calib       ['+str("%.2f" %(ical_c_k))+','+str("%.2f" %(ical_c_r))+','+str("%.2f" %(ical_c_n))+']',linestyle="-")


plt.legend()
plt.title('Basin Average Evapotranspiration',size = 12, weight = 'bold')
plt.ylabel('Evapotranspiration (mm)')
plt.xlabel('Time')
```
