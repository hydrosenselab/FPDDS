```python
import numpy as np
import pandas as pd
import rasterio as rio
import os
pd.set_option('display.max_columns', None)
```

```python
gauge_meta = pd.read_csv('../../final_results/IWM-Validation-EXP002/rawdata/gaugestreamflow/merged/all_year/error_metric_all_year.csv')
```

```python
gauge_meta
```

```python
lat = gauge_meta['Latitude'].values
lon = gauge_meta['Longitude'].values
```

```python
drainagearea = r"/home/civil/phd/cez198628/gdbc/input/GDBC_AS/da_drainagearea.tif"
```

```python
def extractvariable(tifvar):
    with rio.open(tifvar) as src:
        value = [x for x in src.sample(zip(lon, lat))]
        value = [num for elem in value for num in elem]
    return(value)   
```

```python
gauge_meta['Drainage_Area'] = extractvariable(drainagearea)
```

```python
gauge_meta
```

```python
gauge_stats = gauge_meta[['GaugeID','Latitude','Longitude','Drainage_Area']].copy()
```

```python
gauge_stats.to_csv('../../final_results/IWM-Validation-EXP002/rawdata/hydro_signatures/drainage_area.csv')
```
