```python
routedat = xr.open_mfdataset(route_file_list[0:],combine='by_coords')
lsmdat = xr.open_mfdataset(surface_file_list[0:],combine='by_coords')
```

```python
total_streamflow = routedat['Streamflow_tavg']
max_flow = total_streamflow.max()
max_flow.values
```

```python
monthly_average = (total_streamflow.mean(dim='time'))
```

```python
mx = monthly_average.max()
mx.values
```

```python
plt.style.use('seaborn-white')
plt.figure(figsize=(10,9))

#imshow is fast|
im=xr.plot.imshow(monthly_average, 
                  cmap="plasma",vmin=0,add_colorbar=False)
plt.colorbar(im, orientation="vertical")
plt.title('Average Streamflow in July 2013 m3/s')
plt.grid(False)
plt.axis('off')
# ax.set_facecolor('white')
plt.show()
```

```python
total_runoff = lsmdat['Qs_tavg']*86400
average_runoff = (total_runoff.sum(dim='time'))
```

```python
average_runoff
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,12),dpi = 600,constrained_layout=False )
font_size = 12
plt.style.use('seaborn-white')
im=xr.plot.imshow(average_runoff, 
                cmap="jet",add_colorbar=False,vmax=250)
ax.set_title('Total Runoff in July 2013 (mm/month)',size = 18,loc='left')
ax.set_title("",size = 12)
ax.set_xticks([])
ax.set_yticks([])
ax.set_ylabel('')
ax.set_xlabel('')
ax.tick_params(axis='both', labelsize=12)
divider = axes_grid1.make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%', pad=0.05)
cbar = fig.colorbar(im, cax=cax)
plt.savefig('../output/runoff_july13.png',format = 'png', dpi=300)
plt.close
```
